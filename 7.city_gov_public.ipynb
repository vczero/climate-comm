```python
def output_gov_att_for_vis():
    temp = pd.read_csv('./data/news_tj.csv')
    # 便于可视化，百分制
    temp['gov_attention'] = temp['gov_attention'] * 100
    city_profile = pd.read_csv('./data/cities_en_name.csv')
    city_profile = city_profile[['name', 'en_name', 'pro_name', 'pro_en_name']]
    city_profile = city_profile.rename(columns={'name': 'city_name'})

    temp = pd.merge(city_profile, temp)
    temp = temp.sort_values(by='gov_attention', ascending=False)
    temp.to_csv('./QGIS/city_gov_attention.csv', index=None)
    temp.to_csv('./R/city_gov_attention.csv', index=None)

    return temp

output_gov_att_for_vis()
```

## Top 10 cities

```python
def output_top10():
    temp = pd.read_csv('./QGIS/city_gov_attention.csv').head(10)
    city_profile = pd.read_csv('./data/cities_en_name.csv')
    res = city_profile[city_profile.name.isin(temp['city_name'])]

    res.to_csv('./QGIS/city_gov_attention_top10.csv', index=None)
    return temp

output_top10()
```

## Province

```python
def output_province_avg():
    data = pd.read_csv('./R/city_gov_attention.csv')
    province_avg_att = data.groupby('pro_name')['gov_attention'].mean().reset_index()
    province_avg_att.rename(columns={'gov_attention': 'avg_gov_att'}, inplace=True)

    province_avg_att = province_avg_att.sort_values(by='avg_gov_att', ascending=False)
    province_avg_att.to_csv('./QGIS/province_gov_attention.csv', index=None)
    province_avg_att.to_csv('./R/province_gov_attention.csv', index=None)
    return province_avg_att

output_province_avg()
```
