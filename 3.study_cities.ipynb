```python
# step1: 从百度地图的城市code
# txt_file = "./data/bd_cities.txt"
# with open(txt_file, mode="r", encoding="utf-8") as file:
#     raw_data = file.readlines()

# data = []
# for line in raw_data:
#     line = line.strip()
#     if line: 
#         key, value = line.split(":")
#         key = key.strip()
#         value = value.strip().strip('"').strip(',')
#         data.append({"ID": key, "City": value})

# df = pd.DataFrame(data)
# df['City'] = df['City'].str.replace('"', '', regex=False)

bd_names = pd.read_csv('./data/baidu_index_citycode_mapping.csv')
```

```python
# step2: 获取天地图（Tianditu）标准
with open('./QGIS/中国_市_20251108_GS(2024)0650.geojson', 'r', encoding='utf-8') as file:
    data = json.load(file)
names_map = []
for i in data['features']:
    names_map.append(i['properties']['name'])
```

```python
# step3: 使用 Tianditu 过滤百度的城市（因为百度的城市code还相对较旧，没有更新，例如巢湖市现属于合肥市）
study_cities = bd_names[bd_names.name.isin(names_map)]
study_cities = study_cities.rename(columns={'code': 'baidu_city_code'})
study_cities = study_cities.drop_duplicates()
study_cities.to_csv('./data/study_cities.csv', index=None)
len(study_cities)
```
